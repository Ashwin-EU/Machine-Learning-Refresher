# Set up the validation framework

```python
n_val = int(len(df) * 0.2)
n_test = int(len(df) * 0.2)
n_train = len(df) - n_val - n_test
```

```python
idx = np.arange(len(df))
np.random.shuffle(idx)
df_val = df.iloc[idx[:n_val]]
df_test = df.iloc[idx[n_val:n_val+n_test]]
df_train = df.iloc[idx[n_val+n_test:]]
```

```python
df_train.head()
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = np.log1p(df_train.msrp.values)
y_val = np.log1p(df_val.msrp.values)
y_test = np.log1p(df_test.msrp.values)

del df_train["msrp"]
del df_val["msrp"]
del df_test["msrp"]
```

# Train Linear Regression

```python
def train_linear_regression(X, y):
    X = np.array(X)
    y = np.array(y)
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)
    return w_full[0], w_full[1:]
```

```python
df_train.dtypes
```

# Feature Engineering

```python
base = ['engine_hp', 'engine_cylinders', 'highway_mpg', 'city_mpg', 'popularity']
```

```python
def prepare_X(df):
    df = df.copy()
    df['age'] = 2026 - df.year
    features = base + ['age']
    df_num = df[features].fillna(0)
    X = df_num.values
    return X
X_train = prepare_X(df_train)
```

```python
w0, w = train_linear_regression(X_train, y_train)
```

```python
y_pred_train = w0 + X_train.dot(w)
```

```python
X_val = prepare_X(df_val)
y_pred_val = w0 + X_val.dot(w)
```

```python
sns.histplot(y_pred_train, bins=50, color='red', label='predicted')
sns.histplot(y_train, bins=50, color='blue', label='actual')
plt.legend()
```

```python
def rmse(y, y_pred):
    error = y - y_pred
    mse = (error ** 2).mean()
    return np.sqrt(mse)
```

```python
rmse(y_train, y_pred_train)
```

```python
rmse(y_val, y_pred_val)
```
